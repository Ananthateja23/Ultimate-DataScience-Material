```
Histogram:
----------

(1) continuous data
(2) bin = 10 (default)
(3) size = (max-min)/(no of bins)
```

````python
plt.hist(df['Grade'])
````

````python
plt.hist(df['Grade'])
plt.show()
````

````python
plt.hist(df['Grade'],bins=10,range=(0,200))
plt.xticks(list(range(0,200,20)))
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.title("Histogram of Grade of students")
plt.show()
````

```
Boxplot:
-------
(1) use to check outliers
(2) We can check the skewness from the boxplot
	(2.1) If the right side area of box is more then it is right skewed
	(2.2) If the left side area of box is more then it is left skewed
	(2.3) If the right and left area of box are same, then look the whisker to understand the skewness
```

````python
plt.boxplot(df['Grade'])
plt.show()
````

````python
Q1  = df['Grade'].quantile(0.25)
Q3  = df['Grade'].quantile(0.75)
IQR = Q3-Q1
ll  = Q1-(1.5*IQR)
ul  = Q3+(1.5*IQR)
df[(df['Grade']<ll) | (df['Grade']>ul)]
````

```
Scatter plot: (drawn between two continuous variable)
-------------

(1) mark the data points on the graph
(2) use to check --> linearity, direction, strength
(3) xlabel,ylabel, color blue(default), marker "O"(default)
```

````python
plt.scatter(x=df['StudyHours'],y=df['Grade'])
plt.xlabel("StudyHour")
plt.ylabel("Grade")
plt.show()
````

````python
x = np.arange(6)
y = x
z = x**2
````

```
stacked barplot:
----------------
```

````python
company = ["Google","Amazon","Microsoft","Facebook"]
turnover= [90,136,89,112]
profit  = [40,46,34,27]
plt.bar(company,turnover)
plt.bar(company,profit)
plt.show()
````

````python
plt.bar(company,turnover,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
````

````python
company = ["Google","Amazon","Microsoft","Facebook"]
revenue= [90,136,89,27]
profit  = [40,46,34,112]
````

````python
plt.bar([0,1,2,3],revenue,width=0.4,label="revenue")
plt.bar([0.4,1.4,2.4,3.4],profit,width=0.4,label="profit")
plt.xlabel("company")
plt.xticks([0,1,2,3],company)
plt.legend()
plt.show()
````

```
subplot: 
--------

==> creating multiple plots
```

````python
ecommerce = ["Snapdeal","Alibaba","Amazon","Flipkart"]
Q1_profit = [45,100,70,40]
Q2_profit = [40,105,65,45]
Q3_profit = [42,120,72,50]
Q4_profit = [34,115,60,69]
````

````python
# plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.bar(ecommerce,Q1_profit)

plt.subplot(2,2,2)
plt.bar(ecommerce,Q2_profit)

plt.subplot(2,2,3)
plt.bar(ecommerce,Q3_profit)

plt.subplot(2,2,4)
plt.bar(ecommerce,Q4_profit)

plt.show()
````

````python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.bar(ecommerce,Q1_profit)

plt.subplot(2,2,2)
plt.bar(ecommerce,Q2_profit)

plt.subplot(2,2,3)
plt.bar(ecommerce,Q3_profit)

plt.subplot(2,2,4)
plt.bar(ecommerce,Q4_profit)

plt.show()
````
