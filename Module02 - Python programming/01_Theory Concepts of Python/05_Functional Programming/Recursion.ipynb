### ***Recursion***
#### - function call inside same function or function calling itself is called Recursion
#### - A function which follows recursive approach is called recursive function
#### - Every recursive function should have a base condition inorder to stop infinite recursive calls

### ***Advantages***
#### - We can avoid loops by using recursion
#### - number of lines of code can be reduced

### ***Disadvantage***
#### - It takes lot of memory space during execution

---

```python
# print numbers from 1 to 10
def print_num(num):
    print("Value of num : -->",num)
    if num==11:
        return
    print(num)
    print_num(num+1)# function call
    
print_num(1)# function call
```

```python
# print numbers from -1 to -10
def print_num(num):
    print("Value of num : -->",num)
    if num==-11:
        return
    print(num)
    print_num(num-1)# function call
    
print_num(-1)# function call
```

```python
# sum of digits in a number
def sumdig(num):
    print("Value of num : -->",num)
    if num==0:
        return 0
    else:
        return (num%10)+sumdig(num//10)
sumdig(123)
```

```python
# factorial of a number
def factorial(num):
    print("Value of num : -->",num)
    if num==0 or num==1:
        return 1
    return num*factorial(num-1)
factorial(5)
```

```python
# sum of n numbers
def sum_nnumbers(num):
    print("Value of num : -->",num)
    if num==0:
        return 0
    return num+sum_nnumbers(num-1)
sum_nnumbers(5)
```

```python
# fibonacci number
# 0 1 1 2 3 5
# 1 2 3 4 5 6

def fibo(num):
    print("Value of num : -->",num)
    if num==1 or num==2:
        return num-1
    return fibo(num-1)+fibo(num-2)
fibo(6)
```

```python
def func_new(a):
    print("Value of a : -->",a)
    if a<=0:
        return a
    else:
        return func_new(a-1)
ret_val = func_new(5)
print("return val: -->", ret_val)
```

```python
def sum_n_numbers(n):
    if n==0:
        return 0 
    print("n val : -->",n)
    return n+sum_n_numbers(n-1)
ret_val = sum_n_numbers(5)
print(ret_val)
```

```python
def head_recursion(n):
    if n>0:
        head_recursion(n-1) # head_recursion(1-1)
        print("n val: -->", n) # print(1)
# head_recursion(5)--->head_recursion(1)
def tail_recursion(n):
    if n>0:
        print("n val: -->", n)
        tail_recursion(n-1)
```

```python
head_recursion(5)
```

```python
tail_recursion(5)
```

```python
def finite_recursion(n):
    print("val of n: -->", n)
    if n==0:
        return 0 
    return finite_recursion(n-2) # direct recursion 
finite_recursion(6)
```

```python
finite_recursion(7)
```

```python
def count(child_dict, i):
     if i not in child_dict.keys():
         return 1
     ans = 1
     for j in child_dict[i]:
         ans += count(child_dict, j)
     return ans
child_dict = dict()
child_dict[0] = [1,2]
child_dict[1] = [3,4,5]
child_dict[2] = [6,7,8]
print(count(child_dict,0))
```

```python
def fun(D, s1, s2):
    if s1 < s2:
        D[s1], D[s2] = D[s2], D[s1]
        fun(D, s1+1, s2-1)
d = [100,200,5,6,1000,1,10000,12,500]
fun(d, 2, 5)
print(d)
```

```python
# reverse a string
def revstr(substr):
    print("value of substr: --->", substr)
    if len(substr)==1:
        return substr
    return revstr(substr[1:])+substr[0]
revstr("python")
```
